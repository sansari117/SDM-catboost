```python
df = pd.read_csv("/Users/sansari/Desktop/athaliana_urban_nonurban/sdm/1001genomes_lctype_vars.csv", header=0)
len(df.index)
```

```python
df.head()
# Basic properties of the DataFrame
print("Shape of the DataFrame:", df.shape)             # (rows, columns)
print("Number of rows:", len(df))                      # Number of rows
print("Number of columns:", len(df.columns))           # Number of columns
print("Column names:", df.columns.tolist())            # List of column names
print("Data types:\n", df.dtypes)                      # Data type of each column
```

```python
ls_rm = ['Genotype_id', 'vcf_id', 'Country', 'Placename', 'Latitude', 'Longitude', 'Collection', 'genotype_id_combined_for_vis', 'order_id_vcf', 'order_id_vcf_dp3gq20', 'order_id_vcf_dp3gq20maf10', 'snp_2_13144220', 'snp_2_13158845', 'notes', 'lc20021', 'lc20031', 'lc20041', 'lc20051', 'lc20061', 'lc20071', 'lc20081', 'lc20091', 'lc20101', 'lc20111', 'lc20121', 'mode_lc_igbp', 'crop_noncrop', 'lst_dy_max', 'Admixture_Group']

df =df.drop(labels = ls_rm, axis=1)
```

```python
df_set = df.dropna()
df_set['snp_2_13141712'] = df_set['snp_2_13141712'].astype('int')


print(len(df_set.index))

X_set = df_set.drop('snp_2_13141712', axis=1)
y_set = df_set.snp_2_13141712

print(X_set.head())
print(y_set.head())
```

```python
from sklearn.model_selection import train_test_split
from catboost import CatBoostClassifier
X_train_rfe, X_validation_rfe, y_train_rfe, y_validation_rfe = train_test_split(X_set, y_set, train_size=0.75, random_state=42)

test = CatBoostClassifier()

grid = {'learning_rate': [0.01,0.03,0.04],
        'depth': [3,4, 6],
        'l2_leaf_reg': [1, 3, 5,6],
       'iterations':[100,200,300,500]}

grid_search_result = test.grid_search(grid,
                                       X=X_train_rfe,
                                       y=y_train_rfe,
                                       plot=True)
```

```python
grid_search_result['params']
```

```python
to_keep = ["bio_2","bio_4" , "bio_8" ,  "bio_9","bio_13","bio_15","bio_18","bio_19" ,"cfvo" , "clay"        
, "bdod","nitrogen","silt","phh2o","treecover"   
, "gs","elev","wind_mean","nightlst_max", "ndvi_min"    
, "ndvi_max","et_min","et_max","snp_2_13141712"]

df = df[to_keep]

df = df.dropna()
df['snp_2_13141712'] = df['snp_2_13141712'].astype('int')


X = df.drop('snp_2_13141712', axis=1)
y = df.snp_2_13141712

print(X.shape)
print(y.head())
```

```python
from sklearn.model_selection import train_test_split

X_train, X_validation, y_train, y_validation = train_test_split(X, y, train_size=0.75, random_state=42)
```

```python
from catboost import Pool

train_pool = Pool (data = X_train,
                   label = y_train)
                   #cat_features= categorical_features,
                   #weight= normalized_weights)

validation_pool = Pool (data = X_validation,
                        #cat_features=categorical_features,
                        label = y_validation)

print(y_train.value_counts())
```

```python
from catboost import CatBoostRegressor

model = CatBoostRegressor(#iterations = 200,
                          learning_rate= 0.05,
                          depth= 9,
                          l2_leaf_reg= 7,
                          #eval_metric='RMSE',
                          early_stopping_rounds= 50)

model.fit(train_pool,
          eval_set = validation_pool, # becasue of validation set, we have
          verbose=10,
          #use_best_model=True,
          plot = True)

#print('Model is fitted:{}'.format(model.is_fitted()))
#print('Model params:{}'.format(model.get_params()))
print(model.get_best_iteration())
```

```python
raw_pred= model.predict ( X_validation, prediction_type = 'RawFormulaVal')
print(raw_pred)
```

```python
#how well model performs on new dataset

from catboost import CatBoostClassifier, Pool, metrics, cv
from sklearn.metrics import accuracy_score

metrics = model.eval_metrics(data = validation_pool,
                             metrics = ['RMSE'],
                             ntree_start = 0,
                             ntree_end = 0,
                             eval_period = 1, # On which iteration you want to evaluate a metric. Increase to larger value if it takes long to calculate a metric like AUC 
                             plot=True)
```

```python
from catboost import metrics
dir(metrics)
print(model.get_best_score())

#score(train_pool, y_validation)

#from catboost.utils import get_roc_curve
#get_roc_curve(model, 
#              train_pool,
#              plot=True)
```

```python
from sklearn.metrics import explained_variance_score, r2_score, mean_absolute_error

mae = mean_absolute_error(y_validation.values.ravel(), raw_pred)
explained_variance_score(y_validation,raw_pred)
r2 = r2_score(y_validation.values.ravel(), raw_pred)

c = r2_score(y_validation,raw_pred)


print('Mean Absolute Error:', round(mae, 2))
print('R-squared scores:', round(r2, 2))

print('R2:', round(c, 2))


r2_test = model.score(X_validation, y_validation)
r2_train = model.score(X_train, y_train)

print('R2 train:', round(r2_train, 2))
print('R2 test:', round(r2_test, 2))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Permutation importance
from sklearn.inspection import permutation_importance

result = permutation_importance(model, X_validation, y_validation, scoring = 'neg_root_mean_squared_error',n_repeats=10, random_state = 42, n_jobs= 2)


model_importances = pd.Series(result.importances_mean, index=X_validation.columns)

# Plot permutation importance
fig, ax = plt.subplots()
model_importances.plot.bar(yerr=result.importances_std, ax=ax)
ax.set_title("Feature importances using permutation on full model")
ax.set_ylabel("Mean RMSE decrease")
fig.tight_layout()
```

```python
# Plot importance variable

importances = model.get_feature_importance(train_pool,type = 'LossFunctionChange')
model_ranks=pd.Series(importances,index=X_train.columns).sort_values()
model_ranks
#fig, ax = plt.subplots(figsize=(6,10))
ax=model_ranks.plot(kind='barh')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Permutation importance
from sklearn.inspection import permutation_importance

result = permutation_importance(model, X_validation, y_validation, scoring = 'roc_auc',n_repeats=10, random_state = 42, n_jobs= 2)


model_importances = pd.Series(result.importances_mean, index=X_validation.columns)

# Plot permutation importance
fig, ax = plt.subplots()
model_importances.plot.bar(yerr=result.importances_std, ax=ax)
ax.set_title("Feature importances using permutation on full model")
ax.set_ylabel("Mean RMSE decrease")
fig.tight_layout()
```

```python
from sklearn.inspection import permutation_importance

result = permutation_importance(
    model, X_validation, y_validation, n_repeats=10, random_state=42, n_jobs=2, scoring='roc_auc'
)

sorted_importances_idx = result.importances_mean.argsort()
importances = pd.DataFrame(
    result.importances[sorted_importances_idx].T,
    columns=X.columns[sorted_importances_idx],
)
ax = importances.plot.box(vert=False, whis=10)
ax.set_title("Permutation Importances (test set)")
ax.axvline(x=0, color="k", linestyle="--")
ax.set_xlabel("Decrease in importance score (RMSE)")
ax.figure.tight_layout()
```
